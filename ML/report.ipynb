```python
import numpy as np
import cv2
import pandas as pd

import pickle
from matplotlib import pyplot as plt
import os

from skimage.filters import roberts,sobel,scharr, prewitt
from scipy import ndimage as nd
from skimage import feature
```

```python
img=cv2.imread("AddTest/SLA_049.jpg",0)
plt.imshow(img,cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
mask=cv2.imread("AddTestGT/SLA_GT_049.png")
plt.imshow(mask), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
kernel = cv2.getGaborKernel((9,9),4/5. *np.pi,np.pi/4,0.1,0,ktype=cv2.CV_32F)
gabor=cv2.filter2D(img,cv2.CV_8UC3,kernel)
plt.imshow(gabor,cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
canny=feature.canny(img,sigma=3)
plt.imshow(canny,cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
rob=roberts(img)
plt.imshow(rob,cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
sob=sobel(img)
plt.imshow(sob,cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
sch=scharr(img)
plt.imshow(sch,cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
pre=prewitt(img)
plt.imshow(pre,cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
g3=nd.gaussian_filter(img,sigma=3)
plt.imshow(g3,cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
g7=nd.gaussian_filter(img,sigma=7)
plt.imshow(g7,cmap="gray"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

#img = Image.open("AddTest/SLA_049.jpg")
img=cv2.imread("AddTest/SLA_049.jpg")
M = np.asarray(img)

plt.figure(figsize=(12, 6))
plt.subplot(131), plt.xticks([]), plt.yticks([])

plt.imshow(M[:, :, 0], cmap='Blues', vmin=0, vmax=255)
plt.title("Blue Channel")
plt.subplot(132), plt.xticks([]), plt.yticks([])

plt.imshow(M[:, :, 1], cmap='Greens', vmin=0, vmax=255)
plt.title("Green Channel")
plt.subplot(133), plt.xticks([]), plt.yticks([])

plt.imshow(M[:, :, 2], cmap='Reds', vmin=0, vmax=255)
plt.title("Red Channel")

plt.show()
```
